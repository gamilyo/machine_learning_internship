```python
X = pd.DataFrame(X)
```

```python
X.columns = ["Age","EstimatedSalary"]
```

```python
X = pd.concat([X,X_Gender],axis=1)
```

# optimizing all the KNN paramters using GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
K_range = list(range(1,31))
weight_options = ["uniform","distance"]
p = list(range(1,3))
```

```python
param_grid = dict(n_neighbors=K_range,weights=weight_options,p=p)
print(param_grid)
```

```python
import time
start_grid = time.time()
grid = GridSearchCV(Knn,param_grid,cv=10,scoring='accuracy',return_train_score=True)
grid.fit(X,Y)
end_grid = time.time() - start_grid
```

```python
#time needed to run the grid optimizer
print(end_grid)
```

```python
results = pd.DataFrame(grid.cv_results_)[['mean_test_score','mean_train_score','std_test_score','params']]
```

```python
results.head()
```

```python
#printing the best score and paramters
print(grid.best_score_)
print(grid.best_params_)
print(results['mean_test_score'].argmax())
```

```python
grid_test_scores = results['mean_test_score']
grid_train_scores = results['mean_train_score']
```

```python
plt.figure(figsize=(15,10))
plt.plot(grid_test_scores,color="r")
plt.plot(grid_train_scores)
```

# Threshold Tuning for Logistic regression

```python
thr_log = LogisticRegression()
thr_log.fit(X_train,Y_train)
```

```python
pred_y_thr = thr_log.predict(X_test)
```

```python
prob_y = thr_log.predict_proba(X_test)
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
precision, recall, thresholds = precision_recall_curve(Y_test, prob_y[:,1]) 
```

```python
thresholds.shape
```

```python
plt.title("Precision-Recall vs Threshold Chart")
plt.plot(thresholds, precision[: -1], "b--", label="Precision")
plt.plot(thresholds, recall[: -1], "r--", label="Recall")
plt.ylabel("Precision, Recall")
plt.xlabel("Threshold")
plt.legend(loc="lower left")
plt.ylim([0,1])
```

```python
thresold_ranges = np.linspace(0.2,1,50)
```

```python
thr_acc_test = []
for i in thresold_ranges:
    Y_pred = thr_log.predict_proba(X_test)[:,1]
    for j in range(len(Y_pred)) :
        
        if Y_pred[j] >= i:
            Y_pred[j] = 1
        else:
            Y_pred[j] = 0
    thr_acc_test.append(accuracy_score(Y_test,Y_pred))
plt.plot(thresold_ranges,thr_acc_test,color='r')
plt.xlabel("thresold")
plt.ylabel("accurecy_score")
```
